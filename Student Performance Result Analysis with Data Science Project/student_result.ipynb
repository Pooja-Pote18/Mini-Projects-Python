```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
#change weekly study hours column 
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-oct", "5-10")

df.head()
```

```python
#count - gender distribution
```

```python
plt.figure(figsize=(4,4)) 
ax = sns.countplot(data=df,x="Gender") 
ax.bar_label(ax.containers[0]) 
plt.title("Gender Distributed")
```

```python
#from above chart analysed that:the no.female in data is more than no.of male.
```

```python
gb =df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'}) 
print(gb)
```

```python
#Relationship between parent's education & student's score
plt.figure(figsize=(4,4)) 
sns.heatmap(gb,annot=True) 
plt.title("Relationship between parent's education & student's score") 
plt.show()
```

```python
#from above chart we have concluded that education of parents have a good impaction their scores. 
```
